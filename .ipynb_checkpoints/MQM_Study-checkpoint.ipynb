```python
#Cell 1: Imports

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kendalltau, spearmanr

sns.set_theme(style="whitegrid")

print("Libraries loaded. Ready to analyze.")
```

```python
#Cell 2: Load the Data
file_path = '/Users/elizhang/Documents/llm_judge/MQM_Project/wmt-mqm-human-evaluation/newstest2020/ende/mqm_newstest2020_ende.tsv'

file_path = file_path.replace('\\', '') 

try:
    df = pd.read_csv(file_path, sep='\t')
    print(f"Data loaded successfully! Total annotations: {len(df)}")
    display(df.head())
except FileNotFoundError:
    print("Error: File not found. Please verify the absolute path.")
```

```python
#Cell 3: Data Cleaning & Weighting
# Non-translation: Weight 25
# Major: Weight 5
# Minor: Weight 1
# Exception: Minor errors in "Fluency/Punctuation" only get a weight of 0.1

def calculate_weight(row):
    severity = str(row['severity']).lower()
    category = str(row['category']).lower()
    
    # Rule 1: Non-translation is the worst error (Weight 25)
    if severity == 'non-translation':
        return 25
    
    # Rule 2: Major errors (Weight 5)
    if severity == 'major':
        return 5
    
    # Rule 3: Minor errors (Weight 1), BUT...
    if severity == 'minor':
        # ...Exception: Minor Fluency/Punctuation is only 0.1
        if 'punctuation' in category and 'fluency' in category:
            return 0.1
        return 1
    
    # No error or neutral
    return 0

# Apply the function row-by-row
df['error_weight'] = df.apply(calculate_weight, axis=1)

print("Weights assigned based on MQM rules.")
display(df[['severity', 'category', 'error_weight']].head(10))
```

```python
#Cell 4: Grouping & Calculating Scores

#Aggregate error weights
# Segment Score: Sum of all error weights for that specific segment.
# System Score: Average of all segment scores for that system.

# 1. Calculate the total error score per segment
# We group by System, Document, Segment ID, and Rater
segment_scores = df.groupby(['system', 'doc', 'seg_id', 'rater'])['error_weight'].sum().reset_index()

# Rename the column for clarity
segment_scores.rename(columns={'error_weight': 'segment_mqm_score'}, inplace=True)

# 2. Calculate the System-Level Score (Average of segment scores)
# Lower score = Better quality (fewer errors)
system_scores = segment_scores.groupby('system')['segment_mqm_score'].mean().sort_values()

print("--- TOP SYSTEM RANKINGS (Lower Score is Better) ---")
print(system_scores)
```

```python
# Cell 5: Statistical Analysis & Correlations

#rater counts, workload, overlaps, and the Kendall/Spearman correlations between raters

print("--- ANALYSIS REPORT ---")

# 1. Rater Statistics
raters = df['rater'].unique()
num_raters = len(raters)
print(f"Total Raters: {num_raters} (Paper expects 6)")

# 2. Segments per Rater
print("\n--- Workload (Segments per Rater) ---")
segments_per_rater = segment_scores.groupby('rater')['seg_id'].count()
print(segments_per_rater)
# Paper stats: "ranged from 6,830-7,220 for English German"

# 3. Inter-Rater Correlation (Overlapping Segments)
# We need to pivot the data: Rows=Segments, Columns=Raters
pivot_df = segment_scores.pivot_table(index=['system', 'doc', 'seg_id'], 
                                      columns='rater', 
                                      values='segment_mqm_score')

# Count how many segments have overlap (rated by >1 person)
# We count rows where the number of non-NaN values is > 1
overlap_count = pivot_df.dropna(thresh=2).shape[0]
print(f"\nNumber of segments rated by at least 2 raters: {overlap_count}")

# 4. Correlation Matrix (Kendall Tau & Spearman)
# We calculate the correlation between every pair of raters columns
print("\n--- Inter-Rater Agreement (Kendall Tau) ---")
# Kendall is preferred for MQM/Rankings
kendall_matrix = pivot_df.corr(method='kendall')
display(kendall_matrix)

print("\n--- Inter-Rater Agreement (Spearman) ---")
spearman_matrix = pivot_df.corr(method='spearman')
display(spearman_matrix)

# Optional: Visual Heatmap of Agreement
plt.figure(figsize=(8, 6))
sns.heatmap(kendall_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title("Rater Agreement Heatmap (Kendall Tau)")
plt.show()
```
